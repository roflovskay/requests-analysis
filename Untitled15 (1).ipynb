подготовка перед очисткой данных. информация о структуре таблицы, основная статистика, пропуски в столбцах

```python
df.info()
df.describe(include="all")
df.isna().sum()
```

очистка датасета. df.drop_duplicates() - удаляет дубликаты, df.dropna(subset=["Class Name"]) - удаление строк не имеющих категорию, fillna(df["Age"].median()) - заменяет пустую ячейку с возрастом на медиану по всему столбцу,а .astype(int) приводит к целому числу. df["Recommended IND"] = df["Recommended IND"].fillna(0) - отсутствие ркомендации = 0 = не рекомендует. далее переводит текст строки в строку и считает количество символов в отзыве. далее создание групп по возрасту через группировку и if elif and else. df.head() - проверка, показывает первые строки.

```python
df = df.drop_duplicates()
df = df.dropna(subset=["Class Name"])
df["Age"] = df["Age"].fillna(df["Age"].median()).astype(int)
df.head()
```

```python
df["Rating"] = df["Rating"].fillna(df["Rating"].median())
df["Recommended IND"] = df["Recommended IND"].fillna(0)
df["Review Length"] = df["Review Text"].astype(str).apply(len)
df.head()
```

```python
def age_group(x):
    if x < 25:
        return "18-24"
    elif x < 35:
        return "25-34"
    elif x < 50:
        return "35-49"
    else:
        return "50+"
df["Age Group"] = df["Age"].apply(age_group)
df.head()
```

hist - строит гистограмму, ращделенную по ишты на 20 сегментов. далее подписи в заголовке и по х и у

```python
plt.hist(df["Age"], bins=20)
plt.title("возраст покупателей")
plt.xlabel("возраст")
plt.ylabel("количество отзывов")
plt.show()
```

value_counts - считает количество отзывов в каждой категории и благодаря head(10) показывает только 10 первых - самых популярных.plot(kind="bar") - вертикальные столбцы. далее заголовок и подпись

```python
df["Class Name"].value_counts().head(10).plot(kind="bar")
plt.title("топ-10 категорий по количеству отзывов")
plt.ylabel("отзывы")
plt.show()
```

группировка строк по типу одежды,средний рейтинг по каждой категории.sort_values() - сортировка от меньшего к большему. plot(kind="bar")
 - строит bar chart. 

boxplot показывает распределение рейтинга внутри каждой возрастной группы по сегментам из age group

```python
df.boxplot(column="Rating", by="Age Group")
plt.title("рейтинг по возрастным группам")
plt.suptitle("")
plt.xlabel("возрастная группа")
plt.ylabel("рейтинг")
plt.show()
```

scatter - строит точки, где одна точка это один отзыв. размер 10. загооловок и подписи

```python
plt.scatter(df["Review Length"], df["Rating"], s=10)
plt.title("длина отзыва и рейтинг")
plt.xlabel("длина")
plt.ylabel("рейтинг")
plt.show()
```

value_counts() считает колво, сколько людей поставили 0, а сколько 1. plot(kind="bar")  - 2 столбца. xticks - определяет что 0 это  да, а 1 это  нет.


```python
df["Recommended IND"].value_counts().plot(kind="bar")
plt.title("рекомендации")
plt.xticks([0,1], ["рекомендуют", "не рекомендуют"])
plt.ylabel("количество")
plt.show()
```

группировка покупателей по возрасту и средний рейтинг в каждой группе. в виде bar chart


```python
df.groupby("Age Group")["Rating"].mean().plot(kind="bar")
plt.title("рейтинг по возрастным группам")
plt.ylabel("средний рейтинг")
plt.show()
```

pie chart по 5 самым популярным категориям. autopct='%1.1f%%' -  показывает процентаж


```python
df["Class Name"].value_counts().head(5).plot(kind="pie", autopct='%1.1f%%')
plt.title("топ-5 категорий")
plt.ylabel("")
plt.show()
```

corr - подсчитывание кожффицентов, imshow - показывает как картинку - heatmap. и colorbar() добавляет шкалу, где красный = сильная положительная связь, а синий = отрицательная связь 


```python
corr = df[["Age", "Rating", "Review Length", "Recommended IND"]].corr()

plt.imshow(corr, cmap="coolwarm")
plt.colorbar()
plt.xticks(range(len(corr)), corr.columns, rotation=45)
plt.yticks(range(len(corr)), corr.columns)
plt.title("корреляции")
plt.show()
```

 df[df["Rating"] <= 2]  - фильтр отзывов с рейтингов 1 и 2. value_counts()  это количество отзывов 1-2* в возрастных отзывах. отображение результатов в bar chart


```python
neg = df[df["Rating"] <= 2]

neg["Age Group"].value_counts().plot(kind="bar")
plt.title("негативные отзывы по возрастным группам")
plt.ylabel("количество")
plt.show()
```

Инсайты
1. Основная масса покупателей - 30-50 лет, в особенности 38–42, а до 25 лет отзывов почти нет. После 55 активность падает. Аудитория сайта - женщины среднего возраста, а не молодёжь, тогда маркетинг и продукт стоит ориентировать именно на аудиторию 30–45 лет.

2. Рейтинги почти всех категорий держатся в диапазоне 4.0–4.6, с небольшими различиями. Все товары имеют примерно одинаковое качество. 

3. Отзывов у платьев в 4 раза больше чем у большинства других товаров, кроме knit top и blouses. Именно платья формируют основной обьем рейтинга, восприятия и данных.

4. Возраст практически не влияет на оценку. Все возрастные группы ставят примерно одинаковые медианные оценки - на уровне 4–5. Только у молодой групппы есть единичные низкие оценки. Люди разных возрастов одинаково воспринимают продукцию. 

5. Длина отзыва не влияет на рейтинг. Основная масса отзывов - короткие - до 100 символов. Клиенты ставят оценки вне зависимости от подробностей в отзывах.

6. Средний рейтинг всех возрастных групп очень схожий - 4.2-4.4. И молодые покупатели, и более взрослые примерно одинаково оценивают товары. Возраст не влияет на оценку товара, и все позитивные. 

7. Рекомендованных товаров гораздо больше, клиенты доверяют товарам компании. Это доказывают и высокие рейтинги.

8. Топ-5 категорий занимают почти 100% отзывов. Платья и трикотаж дают более 65% отзывов, а остальные мере значимы в аналитике.

9. Больше всего негативных отзывов (1100+) в возрастной группе 35-49. Меньше всего у 18-24. В целом возрастная группа 35-49 самые активные покупатели, и доля негативных у них выражена гораздо сильнее, в 18-24 самих по себе мало.

10. Rating and Recommended IND - очень сильная и логичная корреляция, чем больше оценка, тем сильнее вероятность, что товар будет рекомендован. 
А возраст практически не влияет на оценку клиента. Также возраст не влияет на длину отзыва. Есть небольшая связь с тем, что негативные отзывы более длинные. Возраст не влияет на склонность рекомендовать товар.

Получается самая сильная зависимость между рейтингом и рекомендациями.
